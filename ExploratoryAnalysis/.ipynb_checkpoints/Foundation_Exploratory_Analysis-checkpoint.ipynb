```python
import numpy as np
import pandas as pd
from collections import Counter
from plotnine import *
import plotnine as p9
import os
import warnings
```

```python
#os.getcwd()
os.chdir('/Users/sangeetharamamurthy/Desktop/SephoraFoundationAnalysis/Datasets')
review_db = pd.read_csv('sephora_review_db.csv')
os.chdir('/Users/sangeetharamamurthy/Desktop/SephoraFoundationAnalysis/ExploratoryAnalysis')
```

```python
#remove products we don't want
bad_names = ['Capture Dreamskin Fresh & Perfect Cushion Broad Spectrum SPF 50 Refill', 
             'Double Wear Stay-In-Place Foundation Pump',
             'Chromatic Mix – Oil Base',
             'Teint Idole Ultra Longwear Cushion Foundation SPF 50 Refill',
             'Chromatic Mix – Water Base']

df = review_db[~(review_db['name'].isin(bad_names))]
```

### Overview of the Brands

```python
warnings.filterwarnings("ignore")
(p9.ggplot(df, aes('brand')) +
geom_bar() +
theme(figure_size = (11,8)) +
theme(axis_text_x=element_text(rotation=90, hjust=1)) +
ggtitle('Number of Reviews per Brand') + 
xlab('Brands') + 
ylab('Count'))
```

The most popular brand by reviews is Bare Minerals, and it is not particularly close. All of Bare Minerals' products pull in a total of over 30,000 reviews. The next closest brands are Laura Mercier and MAKE UP FOR EVER. Both brands pull in slightly below 20,000 total reviews. From a foundation standpoint, it makes sense that Bare Minerals is so popular. Their main focus when it comes to what they sell is foundation and mainly other face cosmetics like blush. 

```python
brands = list(df['brand'].unique())
avg_reviews = [round((df[df['brand'] == i]['rating']).mean(),2) for i in brands]
count = df['brand'].value_counts()

brand_df = pd.DataFrame({'brand':brands, 'AverageReview':avg_reviews, 'Count': count}).reset_index(drop=True)
brand_df.head()
```

```python
(p9.ggplot(brand_df, aes(x='brand',y='AverageReview', label = 'AverageReview')) +
geom_point(stat='identity', fill = '#5500AA', size = 8) + 
geom_segment(aes(y=0,
                x = 'brand',
                yend = 'AverageReview',
                xend = 'brand'),
            color = '#5500AA') +
geom_text(color = 'white', size = 6) +
theme(figure_size = (11,15)) +
coord_flip() +
ggtitle('Average Reviews by Brand') +
ylab('Average Rating') +
xlab('Brand Name'))
```

The average rating across all brands is shown in the above plot. The brand with the highest average rating for foundation at Sephora is Givenchy at 4.43, and the brand with the lowest average rating are Anastasia Beverly Hills and DermaDoctor, both of which have average ratings around 3.43. 

Without any other information, this plot would suggest that people should gravitate towards Givenchy foundations because of the higher average rating score. However, I believe this would be silly. When it comes to reviews, I think much of the time, we look at them incorrectly because we fail to account for the total number of reviews we have. For example, a brand that has 10,000 total reviews with an average of 4.2 stars should be worth more than a product that only has 5 reviews with an average of 5 stars due to the larger sample size.
